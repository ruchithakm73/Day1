Machine Learning is mainly divided into three types based on how the model learns from data. 
The difference between them depends on whether the data has labels (answers), no labels, or whether learning happens through rewards and penalties.
Understanding these types helps in choosing the correct approach for solving real-world problems.


 1. Supervised Learning
Supervised Learning is a type of Machine Learning where the data includes labels. 
The model learns the relationship between input data and the known output, so it can predict answers for new data.

2. Unsupervised Learning
Unsupervised Learning is a type of Machine Learning where the data has no labels. 
The model tries to find hidden patterns, structures, or clusters in the data on its own.

3. Reinforcement Learning
Reinforcement Learning is a type of Machine Learning where an agent learns by interacting with an environment.
It receives rewards for correct actions and penalties for wrong actions, improving its decisions over time.


    ## Comparison of Machine Learning Types

| Learning Type | Data Has Labels? | Main Goal | Real-World Examples |
|---------------|------------------|-----------|---------------------|
| Supervised Learning | Yes | Predict outputs or classify data using labeled examples | Spam Detection, House Price Prediction |
| Unsupervised Learning | No | Discover hidden patterns or group similar data | Customer Segmentation, Grouping News Articles |
| Reinforcement Learning | No (uses rewards instead of labels) | Learn the best actions through rewards and penalties | Self-Driving Cars, Game Playing AI |
