```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("internship.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS interns (
    intern_id INTEGER PRIMARY KEY,
    intern_name TEXT,
    track TEXT
)
""")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS mentors (
    mentor_id INTEGER PRIMARY KEY,
    mentor_name TEXT,
    track TEXT
)
""")
```

```python
cursor.executemany("""
INSERT INTO interns (intern_id, intern_name, track)
VALUES (?, ?, ?)
""", [
    (1, "Rahul", "AI"),
    (2, "Rakshith", "Web"),
    (3, "Arjun", "AI"),
    (4, "shamanth", "Data")
])
```

```python
conn.commit()
```

```python
query = """
SELECT i.intern_name, i.track, m.mentor_name
FROM interns i
INNER JOIN mentors m
ON i.track = m.track
"""
```

```python
df = pd.read_sql_query(query, conn)
df
```
